```python
from sklearn.model_selection import train_test_split , cross_val_score
from sklearn.tree import DecisionTreeClassifier
```

```python
X_train , X_test , y_train , y_test = train_test_split(df.drop(columns='severity') , df['severity'] , test_size=0.2 , random_state=42)
```

```python
X_train.head()
```

```python
X_train.dtypes
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import KBinsDiscretizer
```

```python
kbin_cost = KBinsDiscretizer(n_bins=5,encode='ordinal',strategy='kmeans')
```

```python
transformer = ColumnTransformer(transformers=
    [('tnf1',OneHotEncoder(sparse_output=False,drop='first'),['defect_type','defect_location' , 'inspection_method'])
     ,('tnf2',kbin_cost,['repair_cost'])
],remainder='passthrough')
```

```python
X_train_transformed=transformer.fit_transform(X_train)
```

```python
transformer.named_transformers_['tnf2'].bin_edges_
```

```python
   import matplotlib.pyplot as plt
   plt.figure(figsize=(14,4))
   plt.subplot(121)
   plt.hist(X_train['repair_cost'])
   plt.title("Before")
```

```python
X_train_transformed.shape
```

```python
X_train_transformed = X_train_transformed.astype(int)
print("First 5 rows of X_train_transformed after converting to int:")
print(X_train_transformed[:5, :])
print("\nData type of X_train_transformed after conversion:")
print(X_train_transformed.dtype)
```

```python
X_test_transformed=transformer.transform(X_test).astype(int)
X_test_transformed.shape
```

```python
print(X_test_transformed[:5, :])
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y_train_trf= le.fit_transform(y_train)
y_test_trf= le.fit_transform(y_test)
y_test_trf.shape
```

```python
clf = DecisionTreeClassifier()
```

```python
clf.fit(X_train_transformed, y_train_trf)
```

```python
y_pred = clf.predict(X_test_transformed)
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test_trf,y_pred)
```

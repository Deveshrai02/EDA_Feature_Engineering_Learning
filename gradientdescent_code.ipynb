```python
import numpy as np

class GDRegressor:

  def __init__(self , learning_rate, epochs):
    self.learning_rate = learning_rate
    self.epochs = epochs
    self.m = 100
    self.b = -120

  def fit(self, X, y):
    for i in range(self.epochs):
      # Calculate error term (y_predicted - y_actual)
      error = (self.m * X.ravel() + self.b) - y

      # Calculate gradients
      loss_slope_b = 2 * np.sum(error)
      loss_slope_m = 2 * np.sum(error * X.ravel())

      # Update parameters
      self.m = self.m - (self.learning_rate * loss_slope_m)
      self.b = self.b - (self.learning_rate * loss_slope_b)

    print(self.m , self.b)

  def predict(self, X):
    return self.m * X + self.b
```

```python
gd = GDRegressor(0.001 , 100)
gd.fit(X,y)
```
